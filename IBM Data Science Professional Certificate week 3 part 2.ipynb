## 1. Get all Data into data frame and delete records where borough is not assigned

Now let us get data from wikipedia page into a data frame and print the header

```python
wiki = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wiki_page = requests.get(wiki)

wiki_data = pd.read_html(wiki_page.content, header = 0)[0]
df = wiki_data[wiki_data.Neighbourhood != 'Not assigned']
df.reset_index(inplace = True)
df.head()
```

Now lets group by postal code and check neighbourhoods corresonding to those postal codes
